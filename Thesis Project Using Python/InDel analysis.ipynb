```python
def find_insertions(query, subject):
    ins_list = []
    
    ins_start = subject.find('-')
    while ins_start >= 0:
        ins_end = ins_start + 1
        while subject[ins_end] == '-':
            ins_end += 1
            if ins_end == len(subject):
                break
        ins_list.append((query[ins_start:ins_end], ins_start))

        ins_start = subject.find('-', ins_end)
    
    return ins_list

def find_deletions(query, subject):
    return find_insertions(subject, query)

query = 'CATACTATTTTACTAGATATGGGTCTAAACTTTAGCATGCACATTGATCCAATTTAGGCTGAGAGAGGAT'
subject = 'CATACTAT----------ATGGGTCTAAATTTTAGCATGCACATTGATCCAATTTAGGCTGAGAG--GAT'
list_insertions = find_insertions(query, subject)
print(f'insertions: {list_insertions}')
```
