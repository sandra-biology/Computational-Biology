```python
from Bio import SearchIO

def get_hsp_from_file(input_name):
    """
    Function reads XML alignment file and returns HSPs (high-scoring pairs) from the file.
    """
    input_file = SearchIO.read(input_name, 'blast-xml')
    hsp_list = []
    for hit in input_file:
        for hsp in hit.hsps:
            hsp_list.append(hsp)
    return hsp_list

input_name = 'Forrest - Williams 82 Alignment.xml'
list_hsps = get_hsp_from_file(input_name)
print(f'list of HSPs: {list_hsps}')
```

```python
for hsp in list_hsps:
    query_seq = hsp.qseq
    subject_seq = hsp.hseq
    print(f'Query Sequence: {query_seq}')
    print(f'Subject Sequence: {subject_seq}')
```

```python
for hsp in list_hsps:
    query_seq = hsp.query
    subject_seq = hsp.sbjct
    print(f'Query Sequence: {query_seq}')
    print(f'Subject Sequence: {subject_seq}')
```
