```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), x, np.cos(x))

# fig1, ax1 = plt.subplots()
# ax.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)
_ = ax[0].plot(x, np.sin(x))
_ = ax[1].plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x - 0), color="blue")
ax.plot(x, np.sin(x - 1), color="g")
ax.plot(x, np.sin(x - 2), color="0.75")  # grayscale between 0 and 1
ax.plot(x, np.sin(x - 3), color="#FFDD44")  # HEX code between 00 and FF
ax.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  #RGB tuple, values 0 and 1
ax.plot(x, np.sin(x - 5), color="chartreuse")  # all HTML color names supported
```
