```python
def find_all(pattern, string):
    occurrences = []
    index = string.find(pattern)
    while index >= 0:
        occurrences.append(index)
        index = string.find(pattern, index + len(pattern))
    return occurrences

def find_motif(input_name, output_name):
    input_file = open(input_name)
    output_file = open(output_name, 'w')
    n_tests = int(input_file.readline().strip())
    for _ in range(n_tests):
        s = input_file.readline().strip()
        t = input_file.readline().strip()
        # print(f's: {s}, t: {t}')
        
        indices = find_all(t, s)
        print(f'indices list: {indices}')
        indices = [str(x + 1) for x in indices]  # add 1 to count from 1
        print(f'indices string: {indices}')
        output_file.write(' '.join(indices) +'\n')

    output_file.close()
    print(f'wrote output file {output_name}')

input_name = 'input.txt'
# input_name = 'test.txt'
output_name = 'Indices of Motifs.txt'
find_motif(input_name, output_name)
```
