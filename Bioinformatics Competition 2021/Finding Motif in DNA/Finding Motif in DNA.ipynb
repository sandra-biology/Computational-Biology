```python
s = []
t = []
with open("input.txt") as fh:
    fh.readline()
    while True:
        s_str = fh.readline()
        t_str = fh.readline()
        if len(s_str) == 0 or len(t_str) == 0:
            break
        s.append(s_str)
        t.append(t_str)
print(f"s: {s}, t: {t}")
```

```python
print(f"s[:1]: {s[:1]}, t[:5]: {t[:5]}")
s_stripped = [s1.strip() for s1 in s]
t_stripped = [s2.strip() for s2 in t]
print(f"s_stripped[:1]: {s_stripped[:1]}, t_stripped[:5]: {t_stripped[:5]}")
```
